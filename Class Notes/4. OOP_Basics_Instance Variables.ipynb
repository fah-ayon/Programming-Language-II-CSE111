```python
class Student:
    def __init__(self): #constructor/builder/worker method
        print('Hello there!')
        print(self)
        self.name = 'Default' #instance variables
        self.id = 'Jani na'
        self.dept = 'Jani na'
s1 = Student()
print(s1)
print(s1.name, s1.id, s1.dept)
s1.name = 'Harry'
s1.id = 123
s1.dept = 'CSE'
print(s1.name, s1.id, s1.dept)
s2 = Student()
# s2.name = 'Ron'
# s2.id = 129
# s2.dept = 'CSE'
```

```python
class Student:
    def __init__(self, n, i, dep): #constructor/builder/worker method
        #print('Hello there!')
        #print(self)
        self.name = n
        self.id = i
        self.dept = dep

s1 = Student('Harry', 123, 'CSE')
print(s1.name, s1.id, s1.dept)
s2 = Student('Rafeed', 120, 'EEE')
print(s2.name, s2.id, s2.dept)
```
