```python
#Formatting Strings

str1 = 'Hello {}, I am {}. '.format('Steve', 'Emma')
print(str1)

str1 = 'Hello {0}, I am {1}. '.format('Steve', 'Emma') # same but using index. index based on sequence of args in format func
print(str1)

str1 = 'Hello {first_person}, I am {second_person}.'.format( second_person = 'Emma', first_person = 'Steve') #using keyword args
print(str1)
```

```python
#number formatting in strings

s = "Hello {0}, can you lend me {1:d}$?".format("Bob", 100) #{1:d}, 1 is first index, d means decimal values
print(s)

s = "I am {0}, {1}. I have only {2:4.2f}$.".format("sorry", "Alice",50.95876) #2 - index, 4 - how many digits in total, .2f - 2 decimal places
print(s)

print("Binary representation of {0} is {0:b}".format(12)) #'Binary representation of 12 is 1100'
```

```python
#number formatting with allignment

#print("{:<d}".format(20)) #left allignment
print("{:^014.3f}".format(20.1235)) #center allign => total 14 digits with 3 decimal places, blank spaces are filled with 0
print("{:>9d}".format(30)) #right allignment
print("{:>011d}".format(30))
print("{0:<d}|{1:^14.3f}|{2:>9d}".format(20,20.1235,30))
```

```python
#ASCII Code

#Every character is converted to an integer number called the “ASCII code” (0 to 127). then ASCII code is converted to binary numbers
#machine understands binary only
#A (65), Z(90), a(97), z(122)

#convert a character to ascii
print(ord('a')) #use "ord(character)" function
print(ord('A')) #case sensitive

#convert a number to corresponding char
print(chr(122)) #use "chr(int)" function
```

```python
#conversion of text from upper to lower without lower function

text = 'I Love CSE.'
new_text = ''
for char in text:
    if 65<=ord(char)<=90:
        lower_code = ord(char) + 32
        char = chr(lower_code)
    new_text += char
print(new_text)
```
