```python
#Method overloading
#different methods with same mehtod name

class Customer:
    def __init__(self, name, id, phone):
        self.name = name
        self.id = id
        self.phone = phone

#     def addProducts(self): #method1
#         print('No items found')
#     def addProducts(self, item1):#method 2
#         print(item1, 'added')
    def addProducts(self, item1=None, item2=None): #overloaded method
        if item1 == None and item2 == None: #condition to write method1
            print('No items found')
        elif item1 != None and item2 == None:#condition to write method2
            print(item1, 'added')
        else:
            print(item1, 'added')
            print(item2, 'added')
    def showDetails(self):
        info = self.name+' '+str(self.id)+' '+self.phone
        return info
    
c1 = Customer('Alex', 1234, '0176')
c1.addProducts()
c1.addProducts('Mobile')
c1.addProducts('Laptop', 'Mouse')
print('-----------------------')
print(c1.showDetails())
```

```python
#constructor overloading
class Customer:
    def __init__(self, name, id): #constructor1
        self.name = name
        self.id = id
        self.phone = 'Not Found'
    def __init__(self, name, id, phone): #constructor2 is working only
        self.name = name
        self.id = id
        self.phone = phone
    def showDetails(self):
        info = self.name+' '+str(self.id)+' '+self.phone
        return info

c1 = Customer('Alex', 1236)
print(c1.showDetails())
c2 = Customer('John', 1204, '0166')
print(c2.showDetails())
```

```python
#constructor overloading
class Customer:
    def __init__(self, name, id, phone = None): #overloaded constructor
        self.name = name
        self.id = id
        if phone != None:
            self.phone = phone
        else:
            self.phone = 'Not Found'
    def showDetails(self):
        info = self.name+' '+str(self.id)+' '+self.phone
        return info

c1 = Customer('Alex', 1236)
print(c1.showDetails())
c2 = Customer('John', 1204, '0166')
print(c2.showDetails())
```

```python
#constructor overloading with None keyword as default value
class Customer:
    def __init__(self, name = None, id = None, phone = None): #overloaded constructor
        if name == None and id == None and phone == None:# condition to work as constructor1
            self.name = 'Default Name'
            self.id = 'ID Jani Na'
            self.phone = 'Phone Not found'
        elif phone == None:# condition to work as constructor2
            self.name = name
            self.id = id
            self.phone = 'Phone Not Found '
        else:# condition to work as constructor3
            self.name = name
            self.id = id
            self.phone = phone

    def showDetails(self):
        info = self.name+' \n'+str(self.id)+'\n'+self.phone
        return info

c0 = Customer()
print(c0.showDetails(),"\n")
c1 = Customer('Alex', 1236)
#c1.addProducts('Laptop', 'Mouse')
print(c1.showDetails(),'\n')
c2 = Customer('John', 1204, '0166')
print(c2.showDetails())
```

```python
#constructor overloading with *args

class Customer:
    def __init__(self, *args): #overloaded constructor
        if len(args) == 0:
                self.name = 'Default Name'
                self.id = 'ID Jani Na'
                self.phone = 'Phone Not found' 
        elif len(args) == 2: 
            self.name =  args[0]
            self.id = args[1]
            self.phone = 'Phone Not found'
        else: 
            self.name = args[0]
            self.id = args [1]
            self.phone = args[2]    
        
    def showDetails(self):
        info = self.name+' \n'+str(self.id)+'\n'+self.phone
        return info

c0 = Customer()
print(c0.showDetails(),"\n")
c1 = Customer('Alex', 1236)
print(c1.showDetails(),'\n')
c2 = Customer('John', 1204, '0166')
print(c2.showDetails())
```
