```python
class Person: #Parent
    def __init__(self, name):
        self.name = name
        self.identity = 'Human'
    
class Student (Person): #Child(Parent)
    def __init__(self, name, id):
        #Person.__init__(self, name)
        super().__init__(name)
        self.id = id
    def display(self):
        print(self.name, '=>', self.identity)
        print("I am a person")
        
s2 = Student('John', '123')
s2.display()
```

```python
class Student:
    def __init__(self, name, id):
        self.name = name
        self.id = id
class Tutor:
    def __init__(self, tutor_id):
        self.tutor_id = tutor_id

class Student_Tutor(Student, Tutor):
    def __init__(self, name, id, tutor_id):
        super().__init__(name, id)
        Tutor.__init__(self, tutor_id)
    def display(self):
        print(self.name, self.id, self.tutor_id)

s120 = Student_Tutor('Mike', '124', 'T123')
s120.display()
print(s120)
```

```python
class Student:
    def __init__(self, name='Just a student', dept='nothing'):
        self.__name = name #private
        self.__department = dept #private
    def set_department(self, dept):
        self.__department = dept
    def get_name(self):
        return self.__name
    def set_name(self,name):
        self.__name = name
    
    def __str__(self): #overridden method
        return 'Name: '+self.__name+' Department: '+self.__department

class BBA_Student(Student):
    def __init__(self, name = 'default', id = 'jani na', dept = 'BBA'):
        super().__init__(name, dept)
        self.id = id
    
    def __str__(self):
        return super().__str__() + ' ID: ' +str(self.id)
        #return x 
    


print(BBA_Student()) # BBA_Student()
print(BBA_Student('Humpty Dumpty', 23))
print(BBA_Student('Little Bo Peep', 45))
```

```python
class Animal:   
    def __init__(self,sound):
        self.__sound = sound
        
    def makeSound(self):
        return self.__sound
        
class Printer:   
    def printSound(self, a):
        print(a.makeSound())

class Dog(Animal):
    pass
class Cat(Animal):
    pass

d1 = Dog('bark')
c1 = Cat('meow')
a1 = Animal('Animal does not make sound')
pr = Printer()
pr.printSound(a1)
pr.printSound(c1)
pr.printSound(d1)
```
