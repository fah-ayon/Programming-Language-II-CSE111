```python
class Student:
    def display(self):
        print('I am a student')
class Tutor:
    def display(self):
        print('I am a tutor')

class StudentTutor(Tutor, Student):
    def __init__(self, name, id):
        self.name = name
        self.id = id
    def display(self): #overidden method
        super().display()
        #Student.display(self)
        print('Name:', self.name)

st1 = StudentTutor('Jack', 101)
st1.display()
```

```python
class A:
    def __init__(self):
        self._x = 1 #protected

    def methodB(self):
        print('A')
        print(self._x)

class B(A):
    def methodB(self): #override
        print('B')
        print(self._x) #protected var can be accessed by child class
class C:
    def methodB(self, b): 
        print(b._x) #since _x is protected, we should not access it like this (convention)
        pass

b = B()
b.methodB()
c = C()
c.methodB(b)

#print(type(c))
```
