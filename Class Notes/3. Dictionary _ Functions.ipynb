#**Dictionary:**

*   A dictionary is a collection which is ordered (from Python 3.7 version), changeable and indexed.
*   Each item of a dictionary has a key:value pair .

````python
#Creation

#keys must be of any immutable data type - int, float, boolean, string, tuple

new_dict = {} #empty dictionary
#print(new_dict)

dict1 = {4: 'Harry', 1: 'John', 2: 'Carol', 3: 'Moon'} #dictionary of Integer Keys
print(dict1) #ordered

dict2 = {'Name':'John', 'Courses': ['CSE111', 'PHY111'], 1: 'Dhaka', (101, 102): 23.5} #dictionary with mixed keys
#print(dict2)
````

````python
#using built-in dict() function

fruits = dict({1: 'Mango', 2: 'Apple'})
#print(fruits)

# from sequence/iterable having each item as a pair 
#iterable - list, tuple

fruits2 = dict([(1, 'Lemon'), ('abc', ['Orange', 'Banana'])]) #first index of a pair will be key
print(fruits2)
fruits2['abc'][0] = 'Apple'
print(fruits2)
#using keyword arguments.
age_data = dict(Emma = 25, Luke = 23)
print(age_data)
````

````python
#Accessing Elements from Dictionary

#no indexing, need to use KEYS

#using square bracket[key_name]
data = {'Harry': 25, 'Ron': 27, 123: 'bhul info'}
#print(data['Harry']) #put the key in third bracket, check the quote for string

#print(data['123']) #what's wrong??

#using get() method
print(data.get('Ron'))

#difference??
#print(data['Hermione']) #shows error for non-existing keys
#print(data.get('Hermione')) #returns None for non-existing keys
````

````python
#duplicate keys are not possible

data = {'Harry': 25, 'Ron': 27, 123: 'bhul info'}
data ['Harry'] = 45 #replaced old values
data.update({'xyz': 123})

print(data)
````

````python
#Update items

info = {'Name': 'Harry', 'Age': 23, 'School': 'Hogwarts'}
print(info)
info['Name'] = 'Harry Potter' #update existing entry
print(info)
info['Player Type'] = 'Seeker' #new item will be added
print(info)

#using update function
info.update({'Age': 28, 'Place': 'Godric Hollow'})
print(info)
````

````python
#Remove Items

info2 = {'Name': 'Harry', 'Age': 23, 'School': 'Hogwarts', 'House': 'Gryffinder'}

#using pop(key)
print(info2.pop('Age')) #remove a particular item and return value
#print(f"Age of {info2['Name']} is {info2.pop('Age')}") #comment the above line and check it
print(info2)

#using popitem()
print(info2.popitem()) #remove an item from end, return (key,value)

#using del
del info2['School']
print(info2)

#clear() removes all items
info2.clear()
print(info2)

#delete the entire dictionary
del info2
#print(info2) #error - it doesn't exist anymore
````

````python
info2 = {'Name': 'Harry', 'Age': 23, 'School': 'Hogwarts'}
print(info2.keys())
print(list(info2.keys()))
print(info2.values())
print(info2.items()) #returns key-value pair
list1 = list(info2.items())
print(list1) 
````

````python
#Iterating dictionary (Loop):

squares = {1: 1, 3: 9, 5: 25, 7: 49, 9: 81}
for i in squares:
    print(squares[i], end = " ") #just the values

print('\n')

#print values and keys of dictionary
for key, value in squares.items():
    print(key,":", value, end = "\n")

#Assigning value with for loop
new_squares = {}
for x in range(10):
    new_squares[x] = x*x
print(new_squares)

# Assigning values with if conditional (Dictionary Comprehension)
even_squares = {x: x*x for x in range(10) if x % 2 == 0} #{key: value as expression => for iterator in iterable => if condition}
print(even_squares)

#more on dictionary comporehension https://www.programiz.com/python-programming/dictionary-comprehension
````

````python
#Membership Test

# Membership Test for Dictionary Keys
info3 = {'Name': 'Harry', 'Age': 23, 'School': 'Hogwarts'}

print('Name' in info3)
print('House' in info3)
print('Harry' in info3) #Membership test checks for key only,not value
````

````python
#Dictionary Methods

#len(dict)
info4 = {'Name': 'Harry', 'Age': 23, 'School': 'Hogwarts'}
print(len(info4)) #number of items = length


#str(dict)
string1 = str(info4) #Produces a printable string representation of a dictionary
print(string1)
print(type(string1))

#fromkeys(sequence ,v)

list1 = [1,2,3]
val = 10.5
dict2 = dict.fromkeys(list1, val) #Returns a new dictionary with keys from sequence and value equal to v
print(dict2)

#sorted(dictionary)

dict5 = {4:'d', 2:'b', 1: 'a', 3:'c' }
s = sorted(dict5, reverse = True) #sorted list of keys
print(dict5)
print(s)
````

````python
#User-defined Functions

#two types - void & fruitful
#void function doesn't return anything
#fruitful function return some value

'''
x = 10
x = x + 2
x = x*2
x = x + 3
x = x*3
x = x + 10
x = x*10
x = x + 90
x = x*90

print(x) #output = 90000
'''

#To avoid same instructions over and over

def calculate(x, y): #parameters
    x = x + y
    x = x*y
    return x

x = calculate(10, 2) #argumenmts
x = calculate(x, 3)
x = calculate(x, 10)
x = calculate(x, 90)
print(x)
````

````python
#Number of Args

def my_function(name, place):

    print(name+" goes to "+place)

my_function("Alex", "Shopping Complex")
#my_function("Alex") #error - 2 args needed
````

````python
#Unknown Number of Args => *args

def my_function(*kids):
    print(kids)
    print(type(kids))
    print("The first child is "+ kids[0])

my_function("Alice", "Bob", "Carrol")
````

````python
# **kwargs for variable number of keyword arguments
 
def myFun(**kwargs):
    for key, value in kwargs.items():
        print (key, value)
 

myFun(first ='Geeks', mid ='for', last='Geeks') #notice the order
````

````python
#using the *args and **kwargs at the same time

def myFun(*args,**kwargs):
    print("args: ", args)
    print("kwargs: ", kwargs)
 
 
# Now we can use both *args ,**kwargs
# to pass arguments to this function :
myFun('geeks','for','geeks',first="Geeks",mid="for",last="Geeks")
#myFun(first="Geeks", 'geeks','for','geeks',first="Geeks",mid="for",last="Geeks") #error
````

````python
#lambda Function

def average(x, y):
    return (x + y)/2 #single expression
print(average(4, 3))

#using lambda func
print((lambda x, y: (x + y)/2)(4, 3))
````

````python
#recursive function

def recursiveAdd(n):
    if n == 1:
        return 1
    else:
        return n + recursiveAdd(n-1)

print(recursiveAdd(3))
````
