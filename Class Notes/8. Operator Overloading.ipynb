```python
class MyMath2:
    def __init__(self, value):
        self.value = value
    def __add__(self, obj): #self = m1, obj = m2
        return self.value + obj.value # 20 + 30
    def __sub__(self, other):
        return self.value - other.value
    def __eq__ (self, obj):
        return  self.value == obj.value
    def __gt__(self, obj):
        return self.value > obj.value

m1 = MyMath2(20)
m2 = MyMath2(30)
result = m1 + m2 #m1.__add__(m2)
#result = m1.__add__(m2)
print(result)
result2 = m1 - m2 #m1.__sub__(m2)
print(result2)

if m1 == m2: #m1.__eq__(m2)
    print('Equal')
else:
    print('Not Equal')

if m1 > m2: #m1.__gt__(m2)
    print('Greater')
else:
    print('Less or Equal')
```
