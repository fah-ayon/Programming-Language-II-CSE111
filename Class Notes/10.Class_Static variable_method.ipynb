```python
class Student:
    student_counter = 0 #static/class 
    def __init__(self, name, id):
        self.name = name
        self.id = id
        Student.student_counter += 1
        #self.student_counter += 1 #will create another instance variable named student_counter
    def printInfo(self):
        print("Name:", self.name, '\nID:', self.id)
        print(Student.student_counter)
        
#class methods change the state of the class
#works as factory method
#Student object create kora (factory method)
    @classmethod
    def createStudent(cls, info): #factory method
        name, id = info.split('###')
        #print(cls.student_counter)
        return cls(name, id) #Student(name, id)

#static methods are utility method, adds extra functionality to our class
    @staticmethod
    def displayMessage(msg):
        print('Hello', msg, 'Students!')

s1 = Student('Harry', 101)
s1.printInfo()
print(s1.__dict__)
s2 = Student('Ron', 234)
s3 = Student('Neville', 209)
s1.printInfo()
print("----------------------")
s4 = Student.createStudent('Hermione###789')
s4.printInfo()

Student.displayMessage('CSE')
```

```python
from datetime import date
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    # a class method to create a Person object by birthyear.
    @classmethod
    def fromBirthYear(cls, name, year):
        #print(date.today().year)
        age = date.today().year - year
        return cls(name, age)

    # a static method to check if a Person is adult or not.
    @staticmethod
    def isAdult(age): #no self or cls
        return age > 18
        
person1 = Person('John', 21)
person2 = Person.fromBirthYear('Ron', 1996)
print(person1.age)
print(person2.age)
# print the result
print(Person.isAdult(10))
print(person1.isAdult(34))
```
