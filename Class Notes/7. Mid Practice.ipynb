```python
class Author:
    def __init__(self, name = 'Default', *args):
        self.name = name
        self.books = list(args)

    def addBooks(self, *args):
        #print(type(args))
        self.books.extend(args)
        #print(self.books)
    
    def changeName(self, name):
        self.name = name
    def printDetails(self):
        books = '\n'.join(self.books)
        print(f"Author Name: {self.name}\n--------\nList of Books:\n{books}")
        
auth1 = Author('Humayun Ahmed')
auth1.addBooks('Deyal', 'Megher Opor Bari')
auth1.printDetails()
print("===================")
auth2 = Author()
print(auth2.name)
auth2.changeName('Mario Puzo')
auth2.addBooks('The Godfather', 'Omerta', 'The Sicilian')
print("===================")
auth2.printDetails()
print("===================")
auth3 = Author('Paolo Coelho', 'The Alchemist', 'The Fifth Mountain')
auth3.printDetails()
```

```python
#Task 22
class Hotel:
    def __init__(self, name):
        self.hotel = name
        self.stuff = {}
        self.stuffID = 0
        self.guest = {}
        self.guestID = 0
    def addStuff(self, name, age):
        #print(x)
        self.stuffID += 1
        self.stuff.update({self.stuffID :[name, age, '000']})
        print(f"Staff with ID {self.stuffID} is added")
    def getStuffById(self, id):
        return f'Staff ID: {id}\nName: {self.stuff[id][0]}\nAge: {self.stuff[id][1]}\nPhone no.: {self.stuff[id][2]}'
    def addGuest(self, name, age, phone):
        self.guestID += 1
        self.guest.update({self.guestID :[name, age, phone]})
        print(f"Guest with ID {self.guestID} is added")
    def getGuestById(self, id):
        return f'Guest ID: {id}\nName: {self.guest[id][0]}\
        \nAge: {self.guest[id][1]}\nPhone no.: {self.guest[id][2]}'
    def allStaffs(self):
        print('All Staffs:')
        print(f'Number of Staff: {len(self.stuff.keys())}')
        for k, v in self.stuff.items():
            print(f'Staff ID: {k} Name: {v[0]} Age: {v[1]} Phone no: {v[2]}')
    def allGuest(self):
        print('All Guests:')
        print(f'Number of Guest: {len(self.guest.keys())}')
        for k, v in self.guest.items():
            print(f'Guest ID: {k} Name: {v[0]} Age: {v[1]} Phone no: {v[2]}')
#x = 100000
h = Hotel("Lakeshore")
h.addStuff( "Adam", 26)
print("=================================")
print(h.getStuffById(1))
print("=================================")
h.addGuest('Carol',35,'123')
print("=================================")
print(h.getGuestById(1))
print("=================================")
h.addGuest("Diana", 32, '431')
print("=================================")
print(h.getGuestById(2))
print("=================================")
h.allStaffs()
print("=================================")
h.allGuest()
```
